```python
from qiskit import *
```

```python
q = QuantumRegister(3)
```

```python
c = ClassicalRegister(1)
```

```python
cir = QuantumCircuit(q,c)
```

```python
cir.draw('mpl')
```

```python
cir.h(0)
```

```python
cir.cx(0,1)
```

```python
cir.h(2)
```

```python
cir.cx(2,1)
```

```python
cir.x(2)
```

```python
cir.draw('mpl')
```

```python
cir.cx(2,0)
```

```python
cir.x(2)
```

```python
cir.draw('mpl')
```
